```python
import cv2
from utils import get_train_raw_dataset
import matplotlib.pyplot as plt
```

```python
ds = get_train_raw_dataset()
img1 = ds.get_img('SunDataset', 0)
img2 = ds.get_img('TrainDataset', 0)

plt.subplot(121), plt.imshow(img1[:,:,::-1])
plt.subplot(122), plt.imshow(img2[:,:,::-1])
```

```python
import colortrans
import numpy as np
from PIL import Image

# Load data
content = img1
reference = img2

# Transfer colors using different algorithms
output_lhm = colortrans.transfer_lhm(content, reference)
output_pccm = colortrans.transfer_pccm(content, reference)
output_reinhard = colortrans.transfer_reinhard(content, reference)

plt.figure(figsize=(12, 3))
plt.subplot(131), plt.imshow(output_lhm[:,:,::-1])
plt.subplot(132), plt.imshow(output_pccm[:,:,::-1])
plt.subplot(133), plt.imshow(output_reinhard[:,:,::-1])
```
